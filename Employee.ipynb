## Exploratory Data Analysis

```python
df['EducationField'].unique()
```

```python
def segment_model(EducationField):
    if 'Life Sciences' in EducationField:
        return 'Sains'
    elif 'Technical Degree' in EducationField:
        return 'Technical'
    elif 'Medical' in EducationField:
        return 'Health'
    elif 'Marketing' in EducationField:
        return 'Management'
    elif 'Human Resources' in EducationField:
        return 'Psychology'
    else:
        'Other'
        
df['EducationField'] = df['EducationField'].apply(segment_model)
```

```python
plt.figure(figsize=(10,5))
df['EducationField'].value_counts().plot(kind='bar')
```

```python
#list Category to Plot
cat_vars = ['Department', 'EducationField', 'JobRole', 'Gender', 'Age', 'YearsAtCompany']
             
#create figure with subplot
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
axs = axs.ravel()

#create bar plot for each category list
for i, var in enumerate(cat_vars):
    sns.barplot(x=var, y='MonthlyIncome', data=df, ax=axs[i], estimator=np.mean)
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)
    
#adjust spasing between subplots
fig.tight_layout()

#show plot
plt.show()
```

```python
cat_vars = ['JobLevel', 'HourlyRate']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

for i, var in enumerate(cat_vars):
    sns.boxplot(x=var, data=df, ax=axs[i])
    
fig.tight_layout()

plt.show()
```

```python
cat_vars = ['JobLevel', 'HourlyRate']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

for i, var in enumerate(cat_vars):
    sns.violinplot(x=var, data=df, ax=axs[i])
    
fig.tight_layout()

plt.show()
```

```python
num_vars = ['WorkLifeBalance', 'TrainingTimesLastYear', 'TotalWorkingYears', 'TotalWorkingYears']

fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (20,10))
axs = axs.flatten()

for i, var in enumerate(num_vars):
    sns.boxplot(x=var, data=df, ax=axs[i])

fig.tight_layout()

plt.show()
```

```python
num_vars = ['MaritalStatus', 'Gender']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15,15))
axs = axs.flatten()

for i, vars in enumerate(num_vars):
    sns.violinplot(y=var, data=df, x='Gender', ax=axs[i])
    
fig.tight_layout()
plt.show()
```

```python
num_vars = ['MonthlyIncome', 'MonthlyRate'] 

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15,15))
axs = axs.flatten()

for i, vars in enumerate(num_vars):
    sns.boxplot(y=var, data=df, x='Gender', ax=axs[i])
    
fig.tight_layout()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.countplot(df['Gender'], palette='pink')
plt.title('Gender', fontsize=15)
plt.show()
```
