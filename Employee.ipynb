## Exploratory Data Analysis

```python
#list Category to Plot
cat_vars = ['Department', 'EducationField', 'JobRole', 'Gender', 'Age', 'YearsAtCompany']
             
#create figure with subplot
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
axs = axs.ravel()

#create bar plot for each category list
for i, var in enumerate(cat_vars):
    sns.barplot(x=var, y='MonthlyIncome', data=df, ax=axs[i], estimator=np.mean)
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)
    
#adjust spasing between subplots
fig.tight_layout()

#show plot
plt.show()
```

```python
cat_vars = ['JobLevel', 'HourlyRate']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

for i, var in enumerate(cat_vars):
    sns.boxplot(x=var, data=df, ax=axs[i])
    
fig.tight_layout()

plt.show()
```

```python
cat_vars = ['JobLevel', 'HourlyRate']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

for i, var in enumerate(cat_vars):
    sns.violinplot(x=var, data=df, ax=axs[i])
    
fig.tight_layout()

plt.show()
```

```python
num_vars = ['MaritalStatus', 'Gender']

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15,15))
axs = axs.flatten()

for i, vars in enumerate(num_vars):
    sns.violinplot(y=var, data=df, x='Gender', ax=axs[i])
    
fig.tight_layout()
plt.show()
```

## Data Pre-Procesing Part 2

```python
#check missing Value
check_missing = df.isnull().sum()*100 / df.shape[0]
check_missing[check_missing > 0 ].sort_values(ascending=False)
```

## Label Endcoding for Each Object Language

```python
#loop over each colum in the datafreame dtype is 'Object'
for col in df.select_dtypes(include=['object']).columns:
    
    #print the column name and the unique value
    print(f"{col} : {df[col].unique()}")
```

```python
from sklearn import preprocessing

#loop over each colum in the datafreame dtype is 'Object'
for col in df.select_dtypes(include=['object']).columns:
    
    #initialize a labelEncoder obejct
    label_encoder = preprocessing.LabelEncoder()
    
    #fit encoder to the unique values in the column
    label_encoder.fit(df[col].unique())
    
    #transform column using labelEncoder
    df[col] = label_encoder.transform(df[col])
    
    #print the column name and unique encoder value
    print(f"{col} : {df[col].unique()}")
```
